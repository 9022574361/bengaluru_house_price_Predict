```python
import pandas as pd
import numpy as np

# Select only numeric columns
num_cols = df.select_dtypes(include=np.number).columns

# Outlier detection using IQR
outlier_summary = pd.DataFrame(columns=['Column','Outlier_Count'])

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    outlier_summary.loc[len(outlier_summary)] = [col, len(outliers)]

outlier_summary
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def get_accuracy(model,X,y,cv=5):
    return cross_val_score(
        model,
        X=X,
        y=y,
        cv=cv
    )
```

```python
# Change your function to specify the scoring metric
def get_regression_metrics(model, X, y, cv=5):
    # This gives you the R2 scores
    r2_scores = cross_val_score(model, X, y, cv=cv, scoring='r2')
    
    # This gives you the MAE (how many Crores off the prediction is)
    # Note: sklearn returns negative values for MAE in cross_val
    mae_scores = -cross_val_score(model, X, y, cv=cv, scoring='neg_mean_absolute_error')
    
    return r2_scores, mae_scores

r2, mae = get_regression_metrics(lr, ffs_x_train, y_train)
print(f"Average R2: {r2.mean():.4f}")
print(f"Average Error: {mae.mean():.4f} Crores")
```

```python
# ============================================
# XGBOOST HOUSE PRICE MODEL: FINAL PIPELINE
# ============================================
import pandas as pd
import numpy as np
import pickle
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. DEFINE FEATURES AND USE EXISTING y_final
FEATURES = ['total_sqft', 'balcony', 'bedroom_count', 'area_sqft', 'bath_to_room_ratio']

# new_data contains the features, y_final contains the target
X_final = new_data[FEATURES]
# y_final was already correctly aligned in cell [94]
# We use it directly here.

# 2. TRAIN/TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, random_state=42
)

# 3. SCALE THE DATA 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. TRAIN XGBOOST MODEL (With optimized parameters)
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb_model.fit(X_train_scaled, y_train)

# 5. EVALUATION
y_pred = xgb_model.predict(X_test_scaled)
print("--- Final Model Performance ---")
print(f"R2 Score : {r2_score(y_test, y_pred):.4f}")
print(f"MAE      : {mean_absolute_error(y_test, y_pred):.4f} Crores")

# 6. SAVE EVERYTHING (Model + Scaler)
model_package = {
    "model": xgb_model,
    "scaler": scaler,
    "features": FEATURES
}

with open("bengaluru_house_model.pkl", "wb") as f:
    pickle.dump(model_package, f)

print("\n✅ Success: Model and Scaler saved as 'bengaluru_house_model.pkl'")
```
